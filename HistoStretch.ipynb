```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
src = cv2.imread('../data/hawkes.bmp', cv2.IMREAD_GRAYSCALE)
img = cv2.imread('../data/hawkes.bmp')
print(src.shape)
print(img.shape)

plt.figure(figsize=(7, 7))
plt.axis('off')
plt.imshow(img)
plt.title('Origianl Image')
plt.show()
```

```python
# TODO: Histogram display
plt.hist(src.flatten(), range=[0,256], bins=256);
plt.show()
```

```python
# TODO: Histogram Stretching

min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(src)
print('Gmin =', min_val)
print('Gmax =', max_val)
```

```python
dst = ((src-min_val)*255) / (max_val-min_val)
dst = cv2.convertScaleAbs(dst)

cv2.imshow('original', src)
cv2.imshow('Stretching', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
