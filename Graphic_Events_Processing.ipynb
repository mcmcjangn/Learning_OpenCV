```python
#Mouse Event
import numpy as np
import cv2

def onMouse(event , x, y ,flags, param) :
    if event == cv2.EVENT_LBUTTONDOWN:
        if flags & cv2.EVENT_FLAG_SHIFTKEY:
            #Lbutton과 shift키 를 눌렀을 때
            #param[0]은 그 이미지에 그림
            cv2.rectangle(param[0], (x-5, y-5), (x+5, y+5), (255,0,0))
        else:
            cv2.circle(param[0], (x, y), 5, (255,0,0), 3)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(param[0], (x, y), 5, (0,0,255), 3)
    elif event == cv2.EVENT_LBUTTONDBLCLK:
        #더블클릭 했을 때 지우기 (다시 메모리를 할당한거라 좋은 방법은 아님)
        param[0] = np.zeros(param[0].shape, np.uint8) + 255
    
    cv2.imshow("img", param[0])
```

```python
img = np.zeros((512,512,3), np.uint8) + 255
cv2.imshow('img', img)

# 마우스 이벤트핸들러 설정해주는 함수
cv2.setMouseCallback('img', onMouse, [img])

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Mouse Evnet 2
import numpy
import cv2

startx, starty = 0,0
bRect = True
bBtnDwn = False
```

```python
def onMouse(event, x, y, flags, param):
    global startx, starty
    global bRect, bBtnDwn
    
    if event == cv2.EVENT_LBUTTONDOWN:
        startx, starty = x,y
        bRect = True
        bBtnDwn = True
        
    elif event == cv2.EVENT_RBUTTONDOWN:
        startx, starty = x,y
        bRect = False
        bBtnDwn = True
        
    elif event == cv2.EVENT_MOUSEMOVE:
        
        if bRect == True and bBtnDwn == True:
            param[0] = numpy.zeros((512,512,3), numpy.uint8) + 255
            cv2.rectangle(param[0], (startx, starty), (x,y), (255,0,0))
            
        elif bRect == False and bBtnDwn == True :
            sizew, sizeh = x-startx, y-starty
            centerx, centery = (x+startx)//2, (y+starty)//2
            box = ((centerx, centery), (sizew, sizeh), 0)
            param[0] = numpy.zeros((512,512,3), numpy.uint8) + 255
            cv2.ellipse(param[0], box, (0,0,255), 5)
            
        elif event == cv2.EVENT_LBUTTONUP or event == cv2.EVENT_RBUTTONUP:
            bBtnDwn = False
            
        elif event == cv2.EVENT_LBUTTONDBLCLK:
            param[0] = numpy.zeros(param[0].shape, numpy.uint8) + 255
            
        cv2.imshow("img", param[0])
```

```python
img = numpy.zeros((512,512,3), numpy.uint8) + 255
cv2.imshow('img', img)
cv2.setMouseCallback('img', onMouse, [img])
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#key event - 빨간 공 이동
import numpy as np
import cv2

width, height = 512, 512
x, y, R = 256, 256, 50

left=2
right=0
down=1
up=3

direction = right
```

```python
while True:
    key = cv2.waitKeyEx(30)
    if key == 0x1B: #ESC
        break;
    #방향키 방향전환
    elif key == 0x270000: # right
        direction = right
    elif key == 0x280000: # down
        direction = down
    elif key == 0x250000: # left
        direction = left
    elif key == 0x260000: # up
        direction = up

    #방향으로 이동
    if direction == right:
        x+=10
    elif direction == down:
        y += 10
    elif direction == left:
        x -= 10
    else: # 3, up
        y -= 10
        
    # 경계 확인
    if x < R:
        x = R
        direction = right
    if x > width - R:
        x = width - R
        direction = left
    if y < R:
        y = R
        direction = down
    if y > height - R:
        y = height - R
        direction = up
        
    #지우고 그리기
    img = np. zeros((width, height, 3), np.uint8) + 255
    cv2.circle(img, (x,y), R, (0,0,255), -1)
    cv2.imshow('img', img)
    
cv2.destroyAllWindows()
```

```python
#Trackbar event
import numpy as np
import cv2
```

```python
def onChange(pos):
    global img
    #각각의 트랙바는 이름 'R',' G','B' 로 구분한다.
    r = cv2.getTrackbarPos('R', 'img')
    g = cv2.getTrackbarPos('G', 'img')
    b = cv2.getTrackbarPos('B', 'img')
    img[:] = (b, g, r)
    cv2.imshow('img', img)
```

```python
img = np.zeros((512,512,3), np.uint8)#black
cv2.imshow('img',img)

#트랙바 생성
cv2.createTrackbar('R', 'img', 0, 255, onChange)
cv2.createTrackbar('G', 'img', 0, 255, onChange)
cv2.createTrackbar('B', 'img', 0, 255, onChange)

#트랙바 위치 초기화
cv2.setTrackbarPos('R', 'img', 0)
cv2.setTrackbarPos('G', 'img', 0)
cv2.setTrackbarPos('B', 'img', 255)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#event test 트랙바를  통해 이미지 pixel값 각각의 alpha(투명도) 조정
import numpy as np
import cv2

image1 = './data/lena.jpg'
image2 = './data/baboon.jpg'

img1 = 0
img2 = 0
img3 = 0
```

```python
def onChange(pos): # 트랙바 핸들러
    n = cv2.getTrackbarPos('alpha', 'image')
    alpha = n/100;
    #np.uint8 자료형으로 캐스팅 (곱한 값은 float 형이기 떄문에)
    img3 = np.uint8((alpha*img2 + (1-alpha)*img1))
    cv2.imshow('image', img3)
```

```python
img1 = cv2.imread(image1)
img2 = cv2.imread(image2)
img3 = img1.copy()

cv2.imshow('image', img3)
cv2.createTrackbar('alpha', 'image', 0, 100, onChange)

cv2.waitKey()
cv2.destroyAllWindows()
```
